```python
import pyogrio

fire_path = os.path.join(fire_dir, 'Data', 'FPA_FOD_20221014.gdb')

if not 'fire_gdf_raw' in globals():

    print("fire_gdf_raw does not exist. Loading...")

    fire_gdf_raw = pyogrio.read_dataframe(
        fire_path, layer='Fires'
    )

fire_gdf_raw.head()
```

```python
# Name columns and set index for data
fire_gdf = (
    fire_gdf_raw
    [['FOD_ID', 'DISCOVERY_DATE', 'FIRE_SIZE', 'geometry']]
    .set_index('FOD_ID')
)

# Convert the `DISCOVERY_DATE` column to `datetime64`
fire_gdf.DISCOVERY_DATE = pd.to_datetime(fire_gdf.DISCOVERY_DATE)
# Check
# fire_gdf.info()

# Reproject to same as watershed boundary gdf
fire_gdf = fire_gdf.to_crs(wbd_hu2_gdf.crs)

#Call gdf the end of the cell
fire_gdf
```

```python
# Spatially join watershed and fire gdf's

# We want geometry from water boundary dataset so join fire to watershed
# For each row of the fire gdf, we want which watershed hu inside
fire_region_gdf_hu2 = (
    wbd_hu2_gdf
    [['name', 'geometry']]
    .sjoin(fire_gdf, how='inner', predicate='intersects')
)

fire_region_gdf_hu2 = (
    fire_region_gdf_hu2
    .groupby(['name', fire_region_gdf_hu2.DISCOVERY_DATE.dt.year])
    .agg(
        max_fire_size=('FIRE_SIZE', 'max'),
        num_fires=('index_right', 'count'))
)

# print("Number of Fires:", fire_region_gdf.num_fires.sum())
# print("Number of Regions:", len(fire_region_gdf))
# print(round(ans_join.mean(), 2).values)
fire_region_gdf_hu2
```

```python
# Spatially join watershed and fire gdf's

# We want geometry from water boundary dataset so join fire to watershed
# For each row of the fire gdf, we want which watershed hu inside
fire_region_gdf_hu4 = (
    wbd_hu4_gdf
    [['name', 'geometry']]
    .sjoin(fire_gdf, how='inner', predicate='intersects')
)

fire_region_gdf_hu4 = (
    fire_region_gdf_hu4
    .groupby(['name', fire_region_gdf_hu4.DISCOVERY_DATE.dt.year])
    .agg(
        max_fire_size=('FIRE_SIZE', 'max'),
        num_fires=('index_right', 'count'))
)

# print("Number of Fires:", fire_region_gdf.num_fires.sum())
# print("Number of Regions:", len(fire_region_gdf))
# print(round(ans_join.mean(), 2).values)
fire_region_gdf_hu4
```

```python
# Create area column and reproject to EPSG:9822 to get accurate area
# Comes in sq. meters so convert to millions of hectacres
wbd_hu2_gdf['area_ha'] = (
    wbd_hu2_gdf.to_crs(9822).area
    / 10000
    / 1000000
)

# Compute total fires in each watershed
fire_count_df_hu2 = (
    fire_region_gdf_hu2
    .reset_index()
    [['name', 'num_fires']]  # Double square bracket returns dataframe
    .groupby('name')
    .sum()
)

# Add the area and geometry back in
fire_density_gdf_hu2 = (
    wbd_hu2_gdf
    .set_index('name')
    .join(fire_count_df_hu2)
    [['num_fires', 'area_ha', 'geometry']]
)

# Compute fire density
fire_density_gdf_hu2['fire_density_per_ha'] = (
    fire_density_gdf_hu2.num_fires / fire_density_gdf_hu2.area_ha)

# Call column
fire_density_gdf_hu2.fire_density_per_ha
```

```python
# Create area column and reproject to EPSG:9822 to get accurate area
# Comes in sq. meters so convert to millions of hectacres
wbd_hu4_gdf['area_ha'] = (
    wbd_hu4_gdf.to_crs(9822).area
    / 10000
    / 1000000
)

# Compute total fires in each watershed
fire_count_df_hu4 = (
    fire_region_gdf_hu4
    .reset_index()
    [['name', 'num_fires']]  # Double square bracket returns dataframe
    .groupby('name')
    .sum()
)

# Add the area and geometry back in
fire_density_gdf_hu4 = (
    wbd_hu4_gdf
    .set_index('name')
    .join(fire_count_df_hu4)
    [['num_fires', 'area_ha', 'geometry']]
)

# Compute fire density
fire_density_gdf_hu4['fire_density_per_ha'] = (
    fire_density_gdf_hu4.num_fires / fire_density_gdf_hu4.area_ha)

# Call column
fire_density_gdf_hu4.fire_density_per_ha
```

```python
# Plot fire occurrence and size over time for each region


# Set values for ylabels and titles
labels = pd.DataFrame(dict(
    column_name = ['max_fire_size', 'num_fires'],
    ylabel = ['Fire Size (million ha)', 'Number of Fires'],
    title = ['Largest fire on record in each HU2 watershed region',
             'Total number of fires in each HU2 watershed region']))

def fire_plot(region_name, df=fire_region_gdf_hu2, labels=labels):
    # Generate a subplot for each row in the labels
    subplots = []
    # Iterate through the labels row by row
    for i, labs in labels.iterrows():
        # Create subplot
        subplot = (
            df.xs(region_name, level='name')
            [[labs.column_name]]
            .hvplot(
                xlabel='Year',
                ylabel=labs.ylabel,
                title=labs.title,
                width = 800,
                color='red',
                size=2)
            )
        subplots.append(subplot)
        
    # Stack the subplots vertically
    plot = hv.Layout(subplots).cols(1)
    return plot

# Create a dropdown menu to switch between regions
hu2_lineplotbyregion = (
    hv.DynamicMap(
        # The plotting function for the two-panel fire history
        fire_plot,
        # Define the dimension for the dropdown
        kdims=[('region', 'Region')])
    # Add the explicit indexing - region names as a bokeh dimension
    .redim.values(region=(fire_region_gdf_hu2.reset_index().name))
)

hv.save(hu2_lineplotbyregion, 'hu2_lineplotbyregion.html')
```
